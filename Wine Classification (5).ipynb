```python
MAX = 2147483647


class stats:
    def __init__(self, df):
        self.df = df
    
    def mean(self,l:list)->float:
        count = 0
        s = 0
        for i in range(len(l)):
            if (pd.notna(l[i])):
                s = s + l[i]
                count += 1
            else:
                pass
            
        m = s/count
        return m
    
    def median(self,l:list)-> int:
        l.sort()
        cnt = len(l)
        rem = cnt%2
        mid = cnt//2
        if rem==0:
            med = ((mid)+(mid+1))//2
        else:
            med = mid
        return med
    
    def minn(self,l:list)-> float:
        mi = MAX
        for i in range(len(l)):
            if l[i]<mi:
                mi = l[i]
            
        return mi
    
    def maxx(self,l:list) -> float:
        mx = 0.0
        for i in range(len(l)):
            if l[i]>mx:
                mx = l[i]
        
        return mx
    
    def stdv(self,l:list)-> float:
        m = self.mean(l)
        s = 0
        c = 0
        for i in range(len(l)):
                s = s + (l[i]-m)**2
                c += 1
        sd = (s/c)**(1/2)
        
        return sd
    
    def quartile_1(self,l:list)-> float:
        med = self.median(l)
        l.sort()
        q1_cnt = len(l[:med])
        q1_rem = q1_cnt%2
        q1_mid = q1_cnt//2
        if q1_rem==0:
            q1 = ((q1_mid)+(q1_mid+1))//2
        else:
            q1 = q1_mid
            
        return q1
    
    def quartile_3(self,l: list)-> float:
        med = self.median(l)
        l.sort()
        q3_cnt = len(l[med:])
        q3_rem = q3_cnt%2
        q3_mid = (q3_cnt//2)+med
        print(l[q3_mid])
        if q3_rem==0:
            q3 = ((q3_mid)+(q3_mid+1))//2
        else:
            q3 = q3_mid
            
        return q3
    
#     def iqr(self, l:list)-> float:
#         q1 = quartile_1(l)
#         q3 = quartile_3(l)
#         iqrr = q3-q1
#         return iqrr
    
    
    
#     def outliers(self,l:list)-> list:
#         outlist=[]
        
#         nl = [item for item in l if not(math.isnan(item)) == True]

        
#         q1 = self.quartile_1(nl)
#         q3 = self.quartile_3(nl)
#         IQR = q3-q1
#         for item in nl:
#             if (item < q1 - (1.5 * IQR)):
#                 outlist.append(item)
#             elif (item > q3 + (1.5 * IQR)):
#                 outlist.append(item)
#             else:
#                 pass
        
#         return outlist
            
    
    
    def final_dict(self):
        
        cols = self.df.columns.tolist()
        df = pd.DataFrame(index=['mean','stdv','min','q1','q2','q3','max'])
        
        for it in cols:
            stat = []

            l = self.df[it].tolist()
            nl = [item for item in l if not(math.isnan(item)) == True]
            
            
            mea,mi,mx,std,q1,q2,q3 = self.mean(nl),self.minn(nl),self.maxx(nl), self.stdv(nl), self.quartile_1(nl),self.median(nl),self.quartile_3(nl)
            
            
            stat.append(mea)
            stat.append(std)
            stat.append(mi)
            stat.append(nl[q1])
            stat.append(nl[q2])
            stat.append(nl[q3])
            stat.append(mx)

            df[it] = stat
        
        
        return df
#         mea = self.mean(l)

#         return mea
            
            
            
        
```

```python
def impute_by_mean(df):
    cols = df.columns.tolist()
    
    for col in cols:
        mean = 0
        mean = df[col].mean()
        df[col] = df[col].fillna(mean)
        
    return df
```

```python
def impute_by_median(df):
    cols = df.columns.tolist()
    
    for col in cols:
        med = 0
        med = df[col].median()
        df[col] = df[col].fillna(med)
        
    return df
```
